Additional sources of error also include target demographics and marketing budgets which are missing from the dataset but play a major role in sales. Thus, one should only utilize this dataset to get an idea about how a video game is received by the public.


# Webscraping Latest Data from VGChartz

#### The data analysis uses the in the vgsales.csv file as of 6/19/2017. If this file is updated via the webscraping script, it is possible that some of the results no longer match the analysis. Therefore, do not run the script unless you want to obtain updated data.

Notes:  
1) The webscraping script is from https://github.com/GregorUT/vgchartzScrape/blob/master/vgchartzfull.py  
2) Cellphone games are not included in VGChartz data.  
3) Some data is missing in the dataset; we will handle this later.  
4) The webscraping does take a few minutes to complete. 

```python
## Taken from the github link posted above
## DO NOT RUN THIS CELL AS IT WILL UPDATE THE .CSV FILE

from bs4 import BeautifulSoup
import urllib
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

pages = 18
rec_count = 0
rank = []
gname = []
platform = []
year = []
genre = []
publisher = []
sales_na = []
sales_eu = []
sales_jp = []
sales_ot = []
sales_gl = []

urlhead = 'http://www.vgchartz.com/gamedb/?page='
urltail = '&results=1000&name=&platform=&minSales=0.01&publisher=&genre=&sort=GL'

for page in range(1,pages):
	surl = urlhead + str(page) + urltail	
	r = urllib.request.urlopen(surl).read()
	soup = BeautifulSoup(r)
	chart = soup.find("table", class_="chart")
	for row in chart.find_all('tr')[1:]:
		try: 
			col = row.find_all('td')
		
			#extract data into column data
			column_1 = col[0].string.strip()
			column_2 = col[1].string.strip()		
			column_3 = col[2].string.strip()		
			column_4 = col[3].string.strip()		
			column_5 = col[4].string.strip()	
			column_6 = col[5].string.strip()
			column_7 = col[6].string.strip()		
			column_8 = col[7].string.strip()		
			column_9 = col[8].string.strip()		
			column_10 = col[9].string.strip()		
			column_11 = col[10].string.strip()

			#Add Data to columns
			#Adding data only if able to read all of the columns
			rank.append(column_1)
			gname.append(column_2)
			platform.append(column_3)
			year.append(column_4)
			genre.append(column_5)
			publisher.append(column_6)
			sales_na.append(column_7)
			sales_eu.append(column_8)
			sales_jp.append(column_9)
			sales_ot.append(column_10)
			sales_gl.append(column_11)
		
			rec_count += 1
	
		except:
			continue

columns = {'rank': rank, 'name': gname, 'platform': platform, 'year': year, 'genre': genre, 'publisher': publisher, 'NA_Sales':sales_na, 'EU_Sales': sales_eu,'JP_Sales': sales_jp,'Other_Sales':sales_ot, 'Global_Sales':sales_gl }
print (str(rec_count) + " game titles imported")
df = pd.DataFrame(columns)
#df.columns = ['rank','name','platform','year','Genre','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
del df.index.name
df.to_csv("vgsales.csv",sep=",",encoding='utf-8')
```

# Importing the Data

#### General Notes
For this analysis, I will be focusing on Global_Sales as opposed to investigating regional sales. Therefore, I will drop region-specific columns and only maintain Global_Sales which is the sum of all regional sales. Rank denotes the sorted order of the data which is not relevant from a statistical summary perspective. 

#### Global_Sales
This data is in millions of units. The maximum global sales a game has obtained is 82.54 million units while the minimum is 10,000. The average number of sales per game is approximately 500,000 units.

#### Year
The minimum value is 1980 and continues beyond 2017 onwards to 2020. This means that games released before 1980 are not being included in the dataset and that there are even some values beyond today's date. We will investigate this latter item shortly. The year column is also incorrectly formatted as float which will hinder the analysis. Let's fix the year data first:






```python
df = df.drop(['EU_Sales', 'JP_Sales', 'NA_Sales', 'Other_Sales'], 1)
df[df['year'] > 2017]
```

There are only two titles in the population released after 2017. Since it is unclear what the data above is practically, I will assume it is mislabelled and exclude it. This dataset does not include pre-orders.

```python
# Cleaning the data
df = df[df['year'] < 2018]
df = df[pd.notnull(df['year'])]
df['year']=df['year'].astype(int)
```

Although not included in the initial dataset, I'm interested in mapping platforms back to their developers. For example, all Playstation iterations (PS1, PS2, PS3, PSP, PSV, PS4) belong to Sony while all XBox iterations belong to Microsoft. In doing so, I may be able to investigate which companies produce the best platforms as well.

```python
# Mapping the platform to the platform developer
platform_dev = dict.fromkeys(df['platform'])
platform_dev = {
 '2600': 'Atari',
 '3DO': '3D0',
 '3DS': 'Nintendo',
 'DC': 'Sega',
 'DS': 'Nintendo',
 'GB': 'Nintendo',
 'GBA': 'Nintendo',
 'GC': 'Nintendo',
 'GEN': 'Sega',
 'GG': 'Sega',
 'N64': 'Nintendo',
 'NES': 'Nintendo',
 'NG': 'SNK',
 'NS': 'Nintendo',
 'PC': 'Microsoft',
 'PCFX': 'NEC',
 'PS': 'Sony',
 'PS2': 'Sony',
 'PS3': 'Sony',
 'PS4': 'Sony',
 'PSP': 'Sony',
 'PSV': 'Sony',
 'SAT': 'Sega',
 'SCD': 'Sega',
 'SNES': 'Nintendo',
 'TG16': 'HS & NEC',
 'WS': 'Bandai',
 'Wii': 'Nintendo',
 'WiiU': 'Nintendo',
 'X360': 'Microsoft',
 'XB': 'Microsoft',
 'XOne': 'Microsoft'}

df["platform_dev"] = df['platform'].map(platform_dev)

df.head(n=20)
```

From the above, it appears that Nintendo is dominating the market in terms of publishing best-selling games and developing the platform for those games. Nintendo almost exclusively holds the top 15 global-sales spots. However, this doesn't necessarily mean that they are the indisputably the most dominant publisher. Let's look at total global sales:

```python
# Grouping by platform developer and looking at total sales
df[['Global_Sales','platform_dev']].groupby('platform_dev').sum().sort_values(by='Global_Sales', ascending = False).plot.bar(figsize = (10,5))
plt.title("Total Global Sales by Platform Developer (in Millions)");
```

From the chart above, we can observe that despite Nintendo possessing the top best-selling games, they are still beaten out by Sony. This is probably because Sony, as a brand, has more games published across more consoles. The chart also reveals that the biggest companies are Sony, Nintendo and Microsoft in terms of Global Sales. Consequently, it makes sense to focus our analysis on these three companies to gain a better understanding of the market since the competition seems mostly non-exist. 

Next, let's take a glance at how the dataset is distributed:

```python
df_big3 = df[df['platform_dev'].isin(['Sony', 'Microsoft','Nintendo'])]
df_big3['Global_Sales'].hist(figsize = (10,5));
plt.title("Histogram of Sony, Microsoft & Nintendo Global Sales (in Millions)");
plt.ylabel("Frequency");
```

We can see that the vast majority of games are far from best-sellers and have less than 10 million sales. A log-normal transform would transform this into an approximately normal distribution. 

# Correlation between Global Sales & Release Date

One might be inclined to believe that a game released earlier may have more sales because it's been out longer. Let's now investigate whether a correlation exists between a game's release date and its global sales. 

```python
df_big3[['name', 'year']].groupby('year').count().plot.bar(figsize = (10,5));
plt.ylabel('Games Released')
plt.title('Video Game Titles Released by Year')
```

Looks like the data is left-skewed.   

As we entered the 2000s, video game releases began increasing. This could indicate larger public interest in the industry and foreign investors trying to get involved. This sort of trend is common in new industries that demonstrate high potential for profitibility. However, Global Sales don't seem to match the release schedule:

```python
print(df_big3[['year','Global_Sales']].corr())
df.plot.scatter(x = 'year', y = 'Global_Sales', xlim=(1979,2018), ylim=(0,85), figsize=(10,5));
plt.title('Global Sales Scatterplot');
```

From the correlation matrix and the plot, we can observe that there does not appear to be a correlation between a video game's release date and its global sales.  

Although not what one would expect, this result does make sense because in the technology sector, new tech quickly displaces old tech. A game that is revolutionary at the time of its release will be replaced quickly and likely won't see a sales increase. Despite the larger number of well-selling games in the 2000s, we should keep mind that the number of games released during this time period also significantly increased. This in turn increased the industry competitiveness and pushed developers towards consumer interests: leading to bigger, better-received games. Another explanation of the spike in sales is Nintendo's success with the Wii and Motion Sensing technology, which drew untapped markets in 2006 but lost momentum after a few years. 

Ultimately, we fail to reject the null hypothesis because there does not seem to be a consistent trend in the data. A game released in the 1980s could sell just as well as one today and vice-versa. 

# Investigating Relationship Between Global Sales & Genre

Let's now investigate the second hypothesis by visualizing the Average Global Sales (sales / number of games released) by Genre. 

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.title("Average Game Sales (Sales over # of Releases) by Genre")
sns.heatmap(pd.crosstab(df_big3['platform_dev'],df_big3.genre, values = df_big3.Global_Sales, aggfunc = np.mean).fillna(0), annot = True, fmt = '0.2f');
```

Here, we can observe that Nintendo unexpectedly found success with platform and roleplaying games, while Sony and Microsoft found success with Shooters. This is because each of these publishers gained the most sales in these areas relative to the number of games released in them. However, aside from these outliers, there does not seem to be a clear relationship between global sales and genre. 

One should also keep in mind that how VGChartz classified genres is ambiguous here. How does one really distinguish some of these genres? Looking at the data more generally:

```python
df_big3['all'] = 'All Devs'
fig, ax = plt.subplots(figsize=(15,2));
sns.heatmap(pd.crosstab(df_big3['all'],df_big3.genre, values = df_big3.Global_Sales, aggfunc = np.mean).fillna(0), annot = True, fmt = '0.2f');
```

Let's verify whether there is correlation between a game's genre and its global sales with linear regression. 

```python
## Format the genre column as a category and using OLS regression
df_big3["genre"] = df_big3["genre"].astype('category')
result = sm.ols(formula='Global_Sales ~ genre', data=df_big3).fit()
print(result.summary())

## Validate Visually
fig, ax = plt.subplots(figsize=(15,5))
sns.stripplot(x="genre", y="Global_Sales", hue = 'platform_dev', data=df_big3).set(ylim=(0, 90));
```

From the linear regression, we observe an R value of 0.013 which is approximately zero. This indicates that our model was a poor fit and that there likely doesn't exist a directly linear relationship between the sales and genre. 

What this means practically is that although some genres have games with amazing sales numbers, there isn't statistical evidence to suggest that a game of that genre will sell better.
